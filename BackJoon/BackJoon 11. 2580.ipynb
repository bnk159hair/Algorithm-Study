## 새로운 코드

```python
arr = [list(map(int, input().split())) for _ in range(9)]
List = []

for i in range(9):
    for j in range(9):
        if arr[i][j] == 0:
            List.append((i, j))
            


def Row(i, value):
    for k in range(9):
        if arr[i][k] == value:
            return False
    return True
            
            
def Col(j, value):
    for k in range(9):
        if arr[k][j] == value:
            return False
    return True
            
def byandby(i, j, value):
    new_i = i % 3
    new_j = j % 3
        
    for k1 in range(i-new_i, i-new_i+3):
        for k2 in range(j-new_j, j-new_j+3):
            if arr[k1][k2] == value:
                return False
    return True

def sudoku(num):
    if num == len(List):
        for i in range(9):
            print(*arr[i])
        exit(0)
    for i in range(1, 10):
        x, y = List[num]
        
        if Row(x, i) and Col(y, i) and byandby(x, y, i):
            arr[x][y] = i
            sudoku(num+1)
            arr[x][y] = 0
        
    
sudoku(0)

                           
```
