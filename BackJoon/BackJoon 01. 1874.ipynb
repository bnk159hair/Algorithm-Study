```python
import sys

class Stack:
    def __init__(self):
        self.stack = []
    def add(self, num):
        self.stack.append(num)
    def size(self):
        return len(self.stack)
    def empty(self):
        return int(not self.stack)
    def top(self):
        if self.empty():
            return 0
        return self.stack[-1]
    def pop(self):
        if self.empty():
            return 0
        return self.stack.pop()
        

input = sys.stdin.readline
num = int(input())
stack = Stack()

num_arr = [i for i in range(1, num+1)]
result = []

input_arr = [int(input()) for _ in range(num)]

i_num_arr = 0

start = 0
for i in input_arr:
    if i-start>0:
        start = i
        for j in range(i-start):
            stack.add(result[i_num_arr])
            i_num_arr += 1
            result.add("+")
    temp1 = stack.pop()
    if i == temp1:
        result.add("-")
    else:
        print("NO")
    
```
